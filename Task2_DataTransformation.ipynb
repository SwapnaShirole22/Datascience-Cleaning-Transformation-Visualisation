```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Swapna Shirole\Downloads\Iris.csv")
df.head()
```

```python
df_encoded = pd.get_dummies(df, columns=['Species'], drop_first=True)
df_encoded.head()
```

```python
df_encoded['petal_area'] = df['PetalLengthCm'] * df['PetalWidthCm']
df_encoded['sepal_area'] = df['SepalLengthCm'] * df['SepalWidthCm']
df_encoded.head()
```

```python
df_aggregated = df.groupby('Species').agg({
    'SepalLengthCm': 'mean',
    'SepalWidthCm': 'mean',
    'PetalLengthCm': 'mean',
    'PetalWidthCm': 'mean'
}).reset_index()
df_aggregated
```
